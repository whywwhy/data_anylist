```python
import pandas as pd

slave = ['배성은', '김동찬', '조성걸']
jan = [100, 120, 90]
feb = [80, 110, 95]
mar = [90, 100, 85]
apr = [110, 130, 100]
may = [120, 140, 110]
jun = [130, 150, 120]

data = {
    '사원': slave,
    '1월' : jan,
    '2월' : feb,
    '3월' : mar,
    '4월' : apr,
    '5월' : may,
    '6월' : jun
}

df = pd.DataFrame(data)
df.set_index('사원', inplace=True)
df['총판매수'] = df.sum(axis=1)
df['평균판매수'] = df.drop(columns=['총판매수']).mean(axis=1)
df.loc['월간 판매수'] = df.sum()
df.loc['월간 판매평균'] = df.mean()
df
```
