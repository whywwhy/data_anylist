```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
```

```python
X, y = make_moons(n_samples = 100, random_state = 123)
```

```python
plt.scatter(X[y == 0, 0], X[y == 0, 1], color = 'red', marker = '^', alpha = 0.5)
plt.scatter(X[y == 1, 0], X[y == 1, 1], color = 'blue', marker = 'o', alpha = 0.5)

plt.tight_layout()
plt.show
```

```python
from sklearn.decomposition import KernelPCA
import matplotlib.pyplot as plt
```

```python
scikit_kpca = KernelPCA(n_components=2, kernel='rbf', gamma=15)
X_skernpca = scikit_kpca.fit_transform(X)
```

```python
plt.scatter(X_skernpca[y == 0, 0], X_skernpca[y == 0, 1], color='red', marker='^', alpha = 0.5)
plt.scatter(X_skernpca[y == 1, 0], X_skernpca[y == 1, 1], color='blue', marker='o', alpha = 0.5)

plt.xlabel('PC1')
plt.xlabel('PC2')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import KernelPCA
from sklearn.datasets import fetch_california_housing
from sklearn.preprocessing import StandardScaler

# 데이터셋 로드
data = fetch_california_housing()
X = data.data
y = data.target

# 데이터를 표준화
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# KernelPCA 적용
scikit_kpca = KernelPCA(n_components=2, kernel='rbf', gamma=15)
X_skernpca = scikit_kpca.fit_transform(X_scaled)

# 집값을 기준으로 그룹 나누기 (중위값 기준으로 나눔)
y_binary = (y > y.mean()).astype(int)

# 시각화
plt.scatter(X_skernpca[y_binary == 0, 0], X_skernpca[y_binary == 0, 1], color='red', marker='^', alpha=0.5, label='Low House Value')
plt.scatter(X_skernpca[y_binary == 1, 0], X_skernpca[y_binary == 1, 1], color='blue', marker='o', alpha=0.5, label='High House Value')

plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Kernel PCA of California Housing Data')
plt.legend()
plt.tight_layout()
plt.show()
```
