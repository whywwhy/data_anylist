```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
import numpy as np

#150개의 꽃 샘플에서 꽃잎 길이와 꽃잎 너비를 특성 행렬 X에 할당하고 꽃 품종에 해당하는 클래스 레이블을 벡터 y에 할당
```

```python
iris = datasets.load_iris()
X = iris.data[:, [2, 3]]
y = iris.target

print('클래스 레이블:', np.unique(y))

#클래스 레이블: [0 1 2]
```

```python
#처음 본 데이터에서 훈련된 모델 성능을 평가하기 위하 데이터셋을 훈련 데이터셋과 테스트 데이터셋으로 분할

#사이킷런 model_selection 호출하고 모듈의 train_test_split 함수를 이용 -> X와 Y 배열 랜덤하게
# 30는 테스트 데이터(45개) 70은

X_train, X_test, y_train, y_test = train_test_split(
X, y, test_size=0.3, random_state=1, stratify=y)

print('y의 레이블 카운트:', np.bincount(y))

print('y_train의 레이블 카운트:', np.bincount(y_train))

print('y_test의 레이블 카운트:', np.bincount(y_test))
```

```python
#최상 성능 위해 특성 스케일 조정 필요

from sklearn.preprocessing import StandardScaler
# 특성 표준화

sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

```python
#훈련 데이터 표준화 -> 퍼셉트론 모델 훈련

#사이킷런 알고리즘 기본적 OvR(One-versus-Rest) 방식 사용, 다중 분류 (multiclass classification)지원

from sklearn.linear_model import Perceptron

ppn = Perceptron(eta0=0.1, random_state=1)
ppn.fit(X_train_std, y_train)
```

```python
#사이킷 모델 훈련 -> predict로 예측

y_pred = ppn.predict(X_test_std)
print('잘못 분류된 샘플 개수: %d' % (y_test != y_pred).sum())
```

```python
# 라이브러리 metrics 모듈 아레 성능 지표 구현
# ex) 퍼셉트론 분류 정확도 나눈기

from sklearn.metrics import accuracy_score

print('정확도: %.3f' % accuracy_score(y_test, y_pred))
```

```python
#plot_decision_regions 함수를 사용하여 새로운 퍼셉트론 모델의 결정 경계를 그려서 세 개의 붓꽃 샘플을 잘 구분하는지 시각화함

from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt

def plot_decision_regions(X, y, classifier, test_idx=None, resolution=0.02):
  markers = ('s', 'x', 'o', '^', 'v')
  colors = ('red', 'blue','lightgreen', 'gray', 'cyan')
  cmap = ListedColormap(colors[:len(np.unique(y))])
```

```python
# 결정 경계를 그립니다.
x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx1, xx2 = np.meshgrid(np.aragne(x1_min. x1_max. resolution)),
  np.aragne(x2_min, x2_max, resolution))

Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
Z = Z.reshape(xx1.shape)
plt.contourf(xx1, xx2, Z, alpha=0.3, cmap=cmap)
plt.xlim(xx1.min(), xx1.max())
plt.ylim(xx2.min(), xx2.max())
```

```python
for idx, cl in enumerate(np.unique(y)):
  plt.scatter(x=X[y == cl, 0],
    y=X[y == cl, 1],
    alpha=0.8,
    c= colors[idx],
    marker=markers[idx],
    label=cl,
    edgecolor=None if idx==1 else 'black')
```

```python
# 테스트 샘플을 부각하여 그립니다.
if test_idx:
  # 모든 샘플을 그립니다.
  X_test, y_test = X[test_idx, :], y[test_idx]

  plt.scatter(X_test[:, 0],
    X_test[:, 1],
    facecolor='none',
    edgecolor='black',
    alpha=1.0,
    linewidth=1,
    marker='o',
    s=100,
    label='test set')
```

```python
X_combined_std = np.vstack((X_train_std, X_test_std))
y_combined = np.hstack((y_train, y_test))

plot_decision_regions(X=X_combined_std, y=y_combined,

classifier=ppn, test_idx=range(105, 150))

plt.xlabel('petal length [standardized]')
plt.ylabel('petal width [standardized]')
plt.legend(loc='upper left')

plt.tight_layout()
# plt.savefig('images/03_01.png', dpi=300)
plt.show()
```
