```python
import pandas as pd

# 데이터 프레임 합치기 실습 데이터 준비하기
# 첫 번째 데이터 프레임
df1 = pd.DataFrame({
    'key': ['K0', 'K1', 'K2', 'K3'],
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']
})

# 두 번째 데이터 프레임
df2 = pd.DataFrame({
    'key': ['K0', 'K1', 'K2', 'K4'],
    'A': ['C0', 'C1', 'C2', 'C4'],
    'C': ['D0', 'D1', 'D2', 'D4']
})
```

```python
print('### 첫 번째 데이터 프레잉 ###')
print(df1)

print()

print('### 두 번째 데이터 프레임 ###')
print(df2)
```

```python
print("# concat() 세로로 연결 #")
print(pd.concat([df1, df2], axis=0).reset_index(drop=True))
```

```python
print("### concat() 가로로 연결 ###")
print(pd.concat([df1, df2], axis=1))
```

```python
df3 = pd.DataFrame({



'A': ['A0', 'A1', 'A2', 'A3'],

'B': ['B0', 'B1', 'B2', 'B3'],

'C': ['C0', 'C1', 'C2', 'C3'],

'D': ['D0', 'D1', 'D2', 'D3']

})

df4 = pd.DataFrame({



'A': ['A4', 'A5', 'A6', 'A7'],

'B': ['B4', 'B5', 'B6', 'B7'],

'E': ['E4', 'E5', 'E6', 'E7'],

'F': ['F4', 'F5', 'F6', 'F7']

})

print('세')
print(pd.concat([df3, df4], axis=0).reset_index(drop=True))
print("가")
print(pd.concat([df3, df4], axis=1))
```

```python
print(
  pd.merge(df1, df2,
    on='key',
    how='left',
    suffixes=('_left', '_right'))
)
```

```python
df1 = pd.DataFrame({
'key': ['A', 'B', 'C', 'D'],
'value': range(4)
})
df2 = pd.DataFrame({
'key': ['B', 'D', 'D', 'E'],
'value': range(4, 8)
})

print(
  pd.merge(df1, df2,
    on='key',
    how='left',
    suffixes=('_left', '_right'))
)
print('-------------------------------')

print(
  pd.merge(df1, df2,
    on='key',
    how='right',
    suffixes=('_right', '_elft'))
)
print('-------------------------------')

print(
  pd.merge(df1, df2,
    on='key',
    how='outer',
    suffixes=('_outer', '_innter'))
)
print('-------------------------------')

print(
  pd.merge(df1, df2,
    on='key',
    how='inner',
    suffixes=('_inner', '_outer'))
)
```

```python
df5 = pd.DataFrame({
    'Key' : ['c', 'a', 'd', 'b'],
    'A' : [3, 1, 2, 1],
    'B' : [6, 3, 5, 4]
})

df5 = df5.set_index('Key')

print('### 원본 데이터 ###')
print(df5)
```

```python
print('### A열 기준 내림 차순 정리 ###')
print(df5.sort_values(by='A', ascending=False))

print()

print('### A열 오름차순, B열 내림차순 ###')
sorted_df = df5.sort_values(by=['A', 'B'], ascending=[True, False])
print(sorted_df)
```

```python
print('### 인덱스 기준 오름 차순 정렬 ###')
print(df5.sort_index)

print()

print('### 인덱스 기준 내림 차순 정렬 ###')
print(df5.sort_index(ascending=False))
```

```python
print("### 열 'A'의 순위 계산 ###")
df5['A_rank'] = df5['A'].rank(method='min')
print(df5)
```

```python
print('### A열 기준 내림 차순 정리 ###')
print(df5.sort_values(by='A', ascending=False))

print()

print('### A열 오름차순, B열 내림차순 ###')
sorted_df = df5.sort_values(by=['A', 'B'], ascending=[True, False])
print(sorted_df)

print('### 인덱스 기준 오름 차순 정렬 ###')
print(df5.sort_index)

print()

print('### 인덱스 기준 내림 차순 정렬 ###')
print(df5.sort_index(ascending=False))

print("### 열 'A'의 순위 계산 ###")
df5['A_rank'] = df5['A'].rank(method='min')
print(df5)
```
