```python
import matplotlib.pyplot as plt
# 그래프를 사용할 떄 사용하는 matplotlib 라이브러리 불러오기
import pandas as pd
import matplotlib.font_manager as fm
# 폰트 관리자 불러오기 - 한글 폰트를 설정하기 위해 matplotlib.font_manager
fontpath='/usr/share/fonts/trutype/nanum/NaumBarunGothic.ttf'
# 폰트 파일 경로 설정 - 나눔바른고딕체 변수에 저장
font=fm.FontProperties(fname=fontpath, size=9)
# 폰트 속성 설정 - 나눔바른고딕체, 크기 9
plt.rc('font',family='NanumBarunGothic')
#기본 폰트 설정 - 그래프에 표시되는 글자들을 폰트로 사용
```

```python
data = pd.read_csv('/content/drive/MyDrive/2024 창융 ict/연령_및_성별_인구_–_읍면동_20240803133425.csv', thousands=',', encoding='euc-kr')

plt.rcParams['axes.unicode_minus']=True
```

```python
plt.figure(figsize=(10, 8))
plt.barh(data['연령별'], data['총인구_남자(명)'], color = 'blue', label = '남자')
plt.barh(data['연령별'], -data['총인구_여자(명)'], color = 'red', label = '여자')

plt.title('양쪽으로 퍼지는 인구 피라미드')
plt.xlabel('인구 수')
plt.ylabel('연령')
plt.show()
```

```python
plt.figure(figsize=(14, 8))

# Plotting the line graph for men and women
plt.plot(data['연령별'], data['총인구_남자(명)'], 'o-', color = 'blue', label = '남자')
plt.plot(data['연령별'], data['총인구_여자(명)'], 'x-', color = 'red', label = '여자')

plt.title('인구 꺾은선 그래프')
plt.xlabel('연령')
plt.ylabel('인구수')
plt.legend()

plt.show()
```

```python
bar_width = 0.4
index = range(len(data['연령별']))

plt.figure(figsize=(12, 8))
plt.bar([i - bar_width/2 for i in index],data['총인구_남자(명)'], color = 'blue', label = '남자')
plt.bar([i + bar_width/2 for i in index], data['총인구_여자(명)'], color = 'red', label = '여자')

plt.title('그룹화된 막대 그래프')
plt.xlabel('연령')
plt.ylabel('인구수')
plt.xticks(index, data['연령별'], rotation = 45)
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(14, 8))
plt.scatter(data['연령별'], data['총인구_남자(명)'], color = 'blue', label = '남자')
plt.scatter(data['연령별'], data['총인구_여자(명)'], color = 'red', label = '여자')

plt.title('연령대별 인구 수 산점도')
plt.xlabel('연령')
plt.ylabel('인구수')
plt.legend()

plt.xticks(rotation = 45)
plt.show()
```

```python
import seaborn as sns

HeatMapData = [
    [80, 70, 90, 85],
    [60, 80, 75, 70],
    [70, 85, 80, 90],
    [90, 60, 85, 80]
]
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(HeatMapData, annot=True, cmap='PuBuGn',
             xticklabels=['Math', 'Science', 'English', 'Histroy'],
             yticklabels=['Student 1', 'Student 2', 'Student 3', 'Student 4'])

plt.title('학생들의 과목별 성적')
plt.xlable('과목')
plt.ylable('학생')
plt.show()
```
