## **전처리**

1. combine 코드에서 특수문자로 '(주)'가 들어간 부분과 '(사)'가 들어간 부분이 존재하여 삭제

2. 공고명 중 특정 문자 ((), [], /) 등이 존재하여 워드클라우드 제작 시 정확도가 떨어진다 생각되어 공백으로 대체함

3. 또한 FE 개발자가 'F E'라고 뜨는 문제가 발생하여 그 부분을 replace 함

4. 위 와 같이 BE 개발자도 'B E'라고 떠서 백엔드 로 바꿈.

5. 또한 광고성 공고글로 `국비`

```python
import pandas as pd

# 각각의 필터된 CSV 파일 불러오기
df_incruit = pd.read_csv('/content/drive/MyDrive/2024 머신 코스 - 1/크롤링/filtered_incruit_result.csv')
df_incruit2 = pd.read_csv('/content/drive/MyDrive/2024 머신 코스 - 1/크롤링/filtered_incruit_result2.csv')
df_jobkorea1 = pd.read_csv('/content/drive/MyDrive/2024 머신 코스 - 1/크롤링/filtered_jobkorea_result.csv')
df_jobkorea2 = pd.read_csv('/content/drive/MyDrive/2024 머신 코스 - 1/크롤링/filtered_jobkorea_result2.csv')
df_jobkorea3 = pd.read_csv('/content/drive/MyDrive/2024 머신 코스 - 1/크롤링/filtered_jobkorea_result3.csv')
df_saramin = pd.read_csv('/content/drive/MyDrive/2024 머신 코스 - 1/크롤링/filtered_saramin_result.csv')
df_saramin2 = pd.read_csv('/content/drive/MyDrive/2024 머신 코스 - 1/크롤링/filtered_saramin_result2.csv')

# 데이터프레임들을 하나로 합치기
df_combined = pd.concat([df_incruit, df_jobkorea1, df_jobkorea2, df_jobkorea3, df_saramin, df_saramin2], ignore_index=True)

# 회사명과 공고명에서 특정 문자열 제거
df_combined['회사명'] = df_combined['회사명'].replace({'㈜': '', '(주)': '', '(사)': ''}, regex=True)
df_combined['공고명'] = df_combined['공고명'].replace({'㈜': '', '(주)': '', '(사)': ''}, regex=True)

# 공고명과 회사명에서 특정 문자를 공백으로 대체
for char in ['[', ']', '(', ')', '/', ',']:
    df_combined['회사명'] = df_combined['회사명'].str.replace(char, ' ', regex=False)
    df_combined['공고명'] = df_combined['공고명'].str.replace(char, ' ', regex=False)

# 'F E'를 'FE'로 바꾸기
df_combined['회사명'] = df_combined['회사명'].str.replace('F E', 'FE', regex=False)
df_combined['공고명'] = df_combined['공고명'].str.replace('F E', 'FE', regex=False)

# 'B E'를 '백엔드'로 바꾸기
df_combined['회사명'] = df_combined['회사명'].str.replace('B E', '백엔드', regex=False)
df_combined['공고명'] = df_combined['공고명'].str.replace('B E', '백엔드', regex=False)

# '국비지원', '국비교육', '백엔드'가 포함된 행 제거
df_combined = df_combined[~df_combined['공고명'].str.contains('국비지원|국비교육|백엔드', na=False)]

# 특정 키워드가 포함된 공고명에서 해당 키워드를 제거
keywords_to_remove = [
    '경력 채용', '모집', '개발자 모집', '개발자 채용',
    '프론트엔드 개발자', '모집합니다', '운영', '채용공고', '개발', '경력', '개발자', '채용', '자', '및', '공고',' 및', '직', '를 모십니다', '바', '이상'
]

# 공고명에서 특정 키워드 제거
for keyword in keywords_to_remove:
    df_combined['공고명'] = df_combined['공고명'].str.replace(keyword, '', regex=False)

# 학력 정보를 수정하는 함수 정의
def update_education(row):
    if '대졸' in row['학력']:
        return '대졸↑'
    elif '초대졸' in row['학력']:
        return '초대졸↑'
    elif '고졸' in row['학력']:
        return '고졸↑'
    return row['학력']

# 학력 열 업데이트
df_combined['학력'] = df_combined.apply(update_education, axis=1)

# 회사명과 공고명으로 중복 제거, 하나는 남겨놓기
df_combined = df_combined.sort_values(by=['회사명', '공고명']).drop_duplicates(subset=['회사명', '공고명'], keep='first').reset_index(drop=True)

# 결과 확인
display(df_combined)

# 수정된 데이터를 새로운 파일로 저장
df_combined.to_csv('/content/drive/MyDrive/2024 머신 코스 - 1/크롤링/combined_job_results_updated.csv', index=False)
```
